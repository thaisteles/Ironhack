```python
def price(clarity):
    if clarity == 'SI1':
      return 1
    elif clarity == 'VS1':
      return 4
    elif clarity == 'VS2':
      return 3
    elif clarity == 'SI2':
      return 2
    elif clarity == 'VVS2':
      return 5
    elif clarity == 'VVS1':
      return 6
    elif clarity == 'IF':
      return 7
    elif clarity == 'I1':
      return 0
    else:
      return np.nan

price('SI2')
price('I1')
```

```python
df['clarity_number'] = df.clarity.map(price)
```

```python
def price_color(color):
    if color == 'E':
      return 5
    elif color == 'I':
      return 1
    elif color == 'J':
      return 0
    elif color == 'H':
      return 2
    elif color == 'F':
      return 4
    elif color == 'G':
      return 3
    elif color == 'D':
      return 6
    else:
      return np.nan
```

```python
df['color_number'] = df.color.map(price_color)
```
